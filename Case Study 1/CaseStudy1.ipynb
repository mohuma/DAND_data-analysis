```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_style('darkgrid')

df = pd.read_csv('winequality_edited.csv')
```

```python
df.head()
```

```python
color_counts = df.groupby(['color', 'quality']).count()['pH']
color_counts
```

```python
color_totals = df.groupby('color').count()['pH']
color_totals
```

```python
red_proportions = color_counts['red']/color_totals['red']
red_proportions['9'] = 0
red_proportions
```

```python
white_proportions = color_counts['white']/color_totals['white']
white_proportions
```

```python
ind = np.arange(len(red_proportions))
width = 0.35
```

```python
red_bars = plt.bar(ind, red_proportions, width, color='r', alpha=0.7, label = 'Red Wine')
white_bars = plt.bar(ind+width, white_proportions, width, color='w', alpha=0.7, label = 'White Wine')
plt.xlabel('Quality')
plt.ylabel('Proportions')
plt.title('Proportion by Wine Color and Quality')
locations = ind+width/2
labels = ['3', '4', '5', '6', '7', '8', '9']
plt.xticks(locations, labels)
plt.legend()
```
